```python
import folium
from folium.plugins import MarkerCluster, HeatMap
import pandas as pd

# Center map (mean of coordinates)
center_lat = gdf["latitude"].mean()
center_lon = gdf["longitude"].mean()

m = folium.Map(location=[center_lat, center_lon], zoom_start=2, tiles="CartoDB positron")

# --- Clustered points
cluster_fg = folium.FeatureGroup(name="Landslide Points (Clustered)")
cluster = MarkerCluster(name="Cluster").add_to(cluster_fg)

for _, r in gdf.iterrows():
    # Handle fatalities safely for each row
    fatalities = r.get("fatality_count", 0)
    if pd.isna(fatalities):
        fatalities = 0
    fatalities = int(fatalities)

    popup_html = folium.Popup(
        folium.IFrame(
            html=f"""
            <b>{r.get('event_title','(no title)')}</b><br>
            <b>Date:</b> {r.get('event_date')}<br>
            <b>Country:</b> {r.get('country_name','')}<br>
            <b>Trigger:</b> {r.get('landslide_trigger','')}<br>
            <b>Fatalities:</b> {fatalities}<br>
            <a href="{r.get('source_link','')}" target="_blank">Source</a>
            """,
            width=260, height=140
        ),
        max_width=260
    ) 

    folium.CircleMarker(
        location=[r["latitude"], r["longitude"]],
        radius=4,
        color=trigger_color(r.get("landslide_trigger")),  # custom function
        fill=True,
        fill_opacity=0.7,
        popup=popup_html
    ).add_to(cluster)

cluster_fg.add_to(m)

# --- Heatmap for hotspots
heat_fg = folium.FeatureGroup(name="Hotspots (HeatMap)")
heat_data = gdf[["latitude","longitude"]].dropna().values.tolist()   # ✅ drop NaN to avoid errors
HeatMap(heat_data, radius=10, blur=15, max_zoom=6).add_to(heat_fg)
heat_fg.add_to(m)

# --- Convenience: quick “region zoom” markers (India, Nepal, SE Asia)
folium.Marker([22.0, 79.0], tooltip="Zoom: India").add_to(m)
folium.Marker([28.4, 84.1], tooltip="Zoom: Nepal").add_to(m)
folium.Marker([10.0, 105.0], tooltip="Zoom: SE Asia").add_to(m)

# --- Layer control + fullscreen button
folium.LayerControl(collapsed=False).add_to(m)
```

```python
import folium
from folium.plugins import MarkerCluster, HeatMap
import pandas as pd

# ✅ Convert event_date safely to year
df["year"] = pd.to_datetime(
    df["event_date"], format="%m/%d/%Y %I:%M:%S %p", errors="coerce"
).dt.year

# ✅ Keep only valid lat/lon rows
gdf = df.dropna(subset=["latitude", "longitude"]).copy()
gdf = gdf[gdf["landslide_category"] == "landslide"].copy()

# ----------- Country-wise & Year-wise Map -----------
country = "India"   # Change as needed
year = 2017         # Change as needed

# ✅ Filter data for selected country and year
filtered = gdf[(gdf["country_name"] == country) & (gdf["year"] == year)]

if filtered.empty:
    print(f"No data for {country} in {year}")
else:
    # ✅ Center map on average lat/lon
    center_lat = filtered["latitude"].mean()
    center_lon = filtered["longitude"].mean()
    m = folium.Map(location=[center_lat, center_lon], zoom_start=5, tiles="CartoDB positron")

    # ✅ Marker Clustering
    cluster = MarkerCluster().add_to(m)

    def trigger_color(trigger):
        colors = {
            "Rain": "blue",
            "Earthquake": "red",
            "Flooding": "green",
            "Construction": "orange",
            "Mining": "purple"
        }
        return colors.get(trigger, "gray")

    for _, r in filtered.iterrows():
        fatalities = r.get("fatality_count", 0)
        if pd.isna(fatalities):
            fatalities = 0
        try:
            fatalities = int(float(fatalities))
        except Exception:
            fatalities = 0

        popup_html = folium.Popup(
            folium.IFrame(
                html=f"""
                <b>{r.get('event_title','(no title)')}</b><br>
                <b>Date:</b> {r.get('event_date')}<br>
                <b>Country:</b> {r.get('country_name','')}<br>
                <b>Trigger:</b> {r.get('landslide_trigger','')}<br>
                <b>Fatalities:</b> {fatalities}<br>
                """,
                width=260, height=140
            ),
            max_width=260
        )

        folium.CircleMarker(
            location=[r["latitude"], r["longitude"]],
            radius=4,
            color=trigger_color(r.get("landslide_trigger")),
            fill=True,
            fill_opacity=0.7,
            popup=popup_html
        ).add_to(cluster)

    # ✅ Heatmap layer
    heat_data = filtered[["latitude", "longitude"]].dropna().values.tolist()
    HeatMap(heat_data, radius=10, blur=15, max_zoom=6).add_to(m)

    # ✅ Save map
    m.save(f"landslide_map_{country}_{year}.html")
    print(f"Map exported as landslide_map_{country}_{year}.html")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# ✅ Dropdown widgets
country_dropdown = widgets.Dropdown(
    options=sorted(gdf["country_name"].dropna().unique()),
    value="India",
    description="Country:"
)

year_dropdown = widgets.Dropdown(
    options=sorted(gdf["year"].dropna().unique()),
    value=2017,
    description="Year:"
)

button = widgets.Button(description="Generate Map")

output = widgets.Output()

def on_button_click(b):
    with output:
        clear_output()
        country = country_dropdown.value
        year = year_dropdown.value
        filtered = gdf[(gdf["country_name"] == country) & (gdf["year"] == year)]
        
        if filtered.empty:
            print(f"No data for {country} in {year}")
            return
        
        center_lat = filtered["latitude"].mean()
        center_lon = filtered["longitude"].mean()
        m = folium.Map(location=[center_lat, center_lon], zoom_start=5, tiles="CartoDB positron")
        
        cluster = MarkerCluster().add_to(m)
        
        for _, r in filtered.iterrows():
            fatalities = r.get("fatality_count", 0)
            if pd.isna(fatalities):
                fatalities = 0
            try:
                fatalities = int(float(fatalities))
            except:
                fatalities = 0
            
            popup_html = folium.Popup(
                folium.IFrame(
                    html=f"""
                    <b>{r.get('event_title','(no title)')}</b><br>
                    <b>Date:</b> {r.get('event_date')}<br>
                    <b>Country:</b> {r.get('country_name','')}<br>
                    <b>Trigger:</b> {r.get('landslide_trigger','')}<br>
                    <b>Fatalities:</b> {fatalities}<br>
                    """,
                    width=260, height=140
                ),
                max_width=260
            )
            
            folium.CircleMarker(
                location=[r["latitude"], r["longitude"]],
                radius=4,
                color="red",
                fill=True,
                fill_opacity=0.7,
                popup=popup_html
            ).add_to(cluster)

        heat_data = filtered[["latitude", "longitude"]].dropna().values.tolist()
        HeatMap(heat_data, radius=10, blur=15, max_zoom=6).add_to(m)

        display(m)

button.on_click(on_button_click)

# ✅ Show widgets
display(country_dropdown, year_dropdown, button, output)
```
